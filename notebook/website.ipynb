```python
import csv
import pandas as pd
import numpy as np
from collections import defaultdict
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("disease_data.csv",encoding = "cp1252")
```

```python
df.head()
```

```python
data = df.fillna(method='ffill')
```

```python
# Process Disease and Symptom Names
def process_data(data):
    data_list = []
    data_name = data.replace('^','_').split('_')
    n = 1
    for names in data_name:
        if (n % 2 == 0):
            data_list.append(names)
        n += 1
    return data_list
```

```python
# Data Cleanup
disease_list = []
disease_symptom_dict = defaultdict(list)
disease_symptom_count = {}
count = 0

for idx, row in data.iterrows():
    
    # Get the Disease Names
    if (row['Disease'] !="\xc2\xa0") and (row['Disease'] != ""):
        disease = row['Disease']
        disease_list = process_data(data=disease)
        count = row['Count of Disease Occurrence']

    # Get the Symptoms Corresponding to Diseases
    if (row['Symptom'] !="\xc2\xa0") and (row['Symptom'] != ""):
        symptom = row['Symptom']
        symptom_list = process_data(data=symptom)
        for d in disease_list:
            for s in symptom_list:
                disease_symptom_dict[d].append(s)
            disease_symptom_count[d] = count
```

```python
disease_dict = {}
c = 0
for i in disease_symptom_dict.keys():
    disease_dict[i] = c
    c = c+1
```

```python
disease_dict
```

```python
f = open('cleaned_data.csv', 'w')

with f:
    writer = csv.writer(f)
    for key, val in disease_symptom_dict.items():
        for i in range(len(val)):
            writer.writerow([key, val[i], disease_symptom_count[key]])
```

```python
df = pd.read_csv('cleaned_data.csv',encoding = "cp1252")
df.columns = ['disease', 'symptom', 'occurence_count']
df.head()
```

```python
# Remove any rows with empty values
df.replace(float('nan'), np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
# Encode the Labels
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df['symptom'])
print(integer_encoded)
```

```python
# One Hot Encode the Labels
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```

```python
cols = np.asarray(df['symptom'].unique())
cols
```

```python
# Create a new dataframe to save OHE labels
df_ohe = pd.DataFrame(columns = cols)
df_ohe.head()
```

```python
for i in range(len(onehot_encoded)):
    df_ohe.loc[i] = onehot_encoded[i]
```

```python
# Concatenate OHE Labels with the Disease Column
df_disease = df['disease']
df_concat = pd.concat([df_disease,df_ohe], axis=1)
df_concat.head()
```

```python
df_concat.drop_duplicates(keep='first',inplace=True)
cols = df_concat.columns
cols = cols[1:]
df_concat = df_concat.groupby('disease').sum()
df_concat = df_concat.reset_index()
df_concat[:5]
```

```python
data = df_concat
```

```python
data.head()
```

```python
d = data["disease"]
disease = {}
c = 0
for i in d:
    disease[i] = c
    c = c+1
```

```python
data.replace({'disease':disease},inplace = True)
```

```python
Y = data["disease"]
X = data.drop("disease",axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.model_selection import train_test_split
```

```python
clf = RandomForestClassifier(n_estimators = 20)
clf.fit(X,Y)
```

```python
data.head()
```

```python
def Tester(inpt, ExpectedOP):
    ''' Returns True is prediction matches ExpectedOP else false
        Make this run multi-threaded in the future'''
    Y = clf.predict([inpt])
    return Y == ExpectedOP
```

```python
FailCount = 0
Disease_Names = data.disease
# Disease_Names.head()
for i in range(len(data)):
    if( not(Tester(data.iloc[i][1:].to_numpy(),i)) ):
        FailCount += 1
        print("Model failed to predict for the disead {0} at location {1}".format(Disease_Names.iloc[i], data.iloc[i][0]))
    # break
print("Total Failed: {}".format(FailCount))
print("Accuracy: {}".format(((149-FailCount)/149) * 100))
```
